```python
img = imread('tiger-low-contrast.png')
imshow(img)
```

```python
k = round(img.size*0.05)
s = np.sort(img, axis=None)
s = s[k:len(s)-k]
m, M = s[0], s[-1]
```

```python
img = img.astype('float')
img = (img-m)*(255/(M-m))
img = np.clip(img, 0,255)
img = img.astype('uint8')
```

```python
imshow(img)
```
